```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head(10)
```

```python
df2 = df.groupby(['Genre']).agg({'Annual Income (k$)': 'mean'})
df2.columns = ['Income mean']
df2
```

```python
df3 = df.groupby(['Age']).agg({'Annual Income (k$)': 'mean'})
```

```python
df3
```

```python
import seaborn as sns

sns.barplot(x=sorted(set(df['Age'])), y=df3['Annual Income (k$)'])
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot()
ax.bar(sorted(set(df['Age'])), df3['Annual Income (k$)'])
ax.grid()
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
```
