```python
import pandas as pd
```

```python
df = pd.read_csv('Auto Sales data.csv')
df.head(10)
```

```python
df.info()
```

```python
df['ORDERDATE'] = df['ORDERDATE'].astype('datetime64[ns]')
df.head(10)
```

```python
df['ORDERNUMBER'].nunique()
```

```python
df['DEALSIZE'].value_counts()
```

```python
a = df.groupby(['COUNTRY']).agg({'SALES': 'mean'})
a.info()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot()
ax.barh(sorted(set(df['COUNTRY'])), a['SALES'])
```

```python
df['ORDERDATE'].nunique()
```

```python
b = df.groupby('ORDERDATE').agg({'SALES': 'mean'})
```

```python
def f(l):
    n = []
    for i in l:
        if i not in n:
            n.append(i)
    return n


# plt.plot(f(df['ORDERDATE'].sort_values()), b['SALES'])
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot()
ax.plot(f(df['ORDERDATE'].sort_values()), b['SALES'])
```
