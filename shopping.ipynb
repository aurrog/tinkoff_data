```python
import pandas as pd
```

```python
df = pd.read_csv('shopping_trends_updated.csv')
df.head()
```

```python
df.info()
```

```python
df.groupby('Gender').agg({'Purchase Amount (USD)': 'mean'})
```

```python
df.groupby(['Gender', 'Item Purchased']).agg({'Item Purchased': 'count'})
```

```python
a = df.groupby('Item Purchased').agg({'Item Purchased':'count'})
a
```

```python
import matplotlib.pyplot as plt

plt.pie(a['Item Purchased'], labels=sorted(set(df['Item Purchased'])), autopct='%1.1f%%')
```

```python
df.groupby('Item Purchased').agg({'Purchase Amount (USD)': 'mean'})
```

```python
a = df.groupby('Age').agg({'Age': 'count'})

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot()
ax.bar(sorted(set(df['Age'])), a['Age'])
```
